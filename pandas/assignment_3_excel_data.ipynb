```python
file = './hr_dataset.xlsx'

# loading table data which are in multiple sheets into different dataframes
employee = pd.read_excel(file, sheet_name=0)
department = pd.read_excel(file, sheet_name=1)
job_history = pd.read_excel(file, sheet_name=2)
location = pd.read_excel(file, sheet_name=3)
jobs = pd.read_excel(file, sheet_name=4)
```

```python
# 1. Count the NaN values of all the columns of locations file
location.isnull().sum()
```

```python
# 2. Display the name, salary and department number for those employees whose first name ends with the letter 'm'

cols = ['first_name','last_name', 'salary', 'department_id']
employee[cols].loc[employee.first_name.str.endswith('m')]
```

```python
# 3. display the first name, last name, salary and department number
# for those employees whose first name ends with the letter 'd' or 
# 'n' or 's' and also arrange the result in descending order by department id.

cols = ['first_name', 'last_name', 'salary', 'department_id']
result = employee[cols].loc[employee.first_name.str.endswith('d') | employee.first_name.str.endswith('n') | employee.first_name.str.endswith('s')]
result.sort_values(by='department_id', ascending=False)
```

```python
# 4. Find the employees who work either in department 70 or 90.
employee.loc[employee.department_id.isin([70, 90])]
```

```python
# 5. Display the first, last name, salary and department number 
# for those employees who hold a letter n as a 3rd character 
# in their first name. 

cols = ['first_name', 'last_name', 'salary', 'department_id']
employee[cols].loc[employee.first_name.str[2] == 'n']
```

```python
# 6. Display the first name, job id, salary and department for 
# those employees not working in the departments 50,30 and 80

cols = ['first_name', 'salary', 'job_id', 'department_id']
employee[cols].loc[~employee.department_id.isin([30, 50, 80])]
```

```python
# 7. Display the ID for those employees who did two or more jobs in the past.

ids = job_history.groupby('employee_id').job_id.size().rename('count')
ids = ids.reset_index()

ids.employee_id.loc[ids['count'] >= 2]
```

```python
# 8. Calculate minimum, maximum and mean salary from employees file

min_salary = employee.salary.min()
max_salary = employee.salary.max()
mean_salary = employee.salary.mean()
print(f'Min: {min_salary}, Max: {max_salary}, Mean: {mean_salary}')
```

```python
# 9. Display the first and last name and date of joining of the 
# employees who is either Sales Representative or Sales Man.

cols = ['first_name', 'last_name', 'hire_date']
job_list = ['SA_REP', 'SA_MAN']
employee[cols].loc[employee.job_id.isin(job_list)]
```

```python
# 10. display the first and last name, and department number for 
# all employees whose last name is "McEwen".

cols = ['first_name', 'last_name', 'department_id']
employee[cols].loc[employee.last_name == 'McEwen']
```

```python
# 11. create and display a boolean series, where True for not null 
# and False for null values or missing values in state_province 
# column of locations file.

location.state_province.notnull()
```

```python
# 12. Compute the average salary earned by each department 
# with department name

avg_salary_by_dept = employee.groupby('department_id').salary.mean().rename('avg_salary')

avg_salary_by_dept = pd.DataFrame(avg_salary_by_dept)
joined = avg_salary_by_dept.join(department.set_index('department_id'), how='outer')
joined = joined.fillna(0)
joined[['department_name', 'avg_salary']]
```

```python
# 13. Compute the number of departments located in Seattle city.

# we need location and department dataframes

location_no_depts = department.groupby('location_id').size().rename('num_depts')
location_no_depts = pd.DataFrame(location_no_depts)
location_no_depts = location_no_depts.reset_index()

combined = location_no_depts.merge(location, on='location_id')
combined.loc[combined.city == 'Seattle', ['city', 'num_depts']]
```

```python
# 14. What is the average salary of job Finance Manager

job_title = ['Finance Manager']
combined = employee.merge(jobs, on='job_id')
combined = combined.groupby(['job_title']).salary.mean().rename('avg_salary')
combined.loc[job_title]
```

```python
# 15. You are offered a position either as Sales Representative 
# or Marketing Representative. You want to join a position with 
# a smaller number of employees. Which position will you join?

combined = employee.merge(jobs, on='job_id')
job_counts = combined.groupby('job_title').size()
n_sales_rep = job_counts.loc['Sales Representative']
n_marketing_rep = job_counts.loc['Marketing Representative']

# select a job with less num of employees
print("I will take the job of ", end='')
print('Sales Representative' if n_sales_rep < n_marketing_rep else 'Marketing Representative')
```
