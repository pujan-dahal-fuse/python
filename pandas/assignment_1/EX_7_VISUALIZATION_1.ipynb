### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import seaborn as sns

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep='\t')
```

### Step 4. See the first 10 entries

```python
chipo.head(10)
```

### Step 5. Create a histogram of the top 5 items bought

```python
items_total_quantity = chipo.groupby('item_name').quantity.sum()
top5 = items_total_quantity.sort_values(ascending=False).head(5)
top5 = pd.DataFrame(top5.reset_index())

sns.set(style='darkgrid')
plt = sns.barplot(data=top5, x='item_name', y='quantity')
plt.set_xticklabels(top5.item_name, rotation=25)
```

### Step 6. Create a scatterplot with the number of items orderered per order price
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
# strip dollar sign from item_price
# chipo.item_price = chipo.item_price.map(lambda x: float(x.strip('$')))

quantity_per_item_price = chipo.groupby('item_price').quantity.sum()
df = pd.DataFrame(quantity_per_item_price.reset_index())

sns.scatterplot(data=df, x='item_price', y='quantity', s=50)
```
