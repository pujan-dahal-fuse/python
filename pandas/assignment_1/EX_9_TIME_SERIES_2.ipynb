### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
import pandas_datareader as pdr
import datetime as dt
import os
```

### Step 2. Create your time range (start and end variables). The start date should be 01/01/2015 and the end should today (whatever your today is).

```python
start_date = dt.date(2015, 1, 1)
end_date = dt.date(2022, 9, 22)
```

### Step 4. Use Pandas Datarader to read the daily time series for the Apple stock (ticker symbol AAPL) between 01/01/2015 and today, assign it to df_apple and print it.

```python
ALPHAVANTAGE_API_KEY = 'K15GY8QOVBFUBRW5'

apple_df = pdr.data.DataReader('AAPL', 'av-daily', start=start_date, end=end_date, api_key=ALPHAVANTAGE_API_KEY)

apple_df
```

### Step 5. Add a new column "stock" to the dataframe and add the ticker symbol

```python
apple_df.insert(5, 'stock', 'AAPL')
apple_df
```

### Step 6. Repeat the two previous steps for a few other stocks, always creating a new dataframe: Tesla, IBM and Microsoft. (Ticker symbols TSLA, IBM and MSFT.)

```python
tesla_df = pdr.data.DataReader('TSLA', 'av-daily', start=start_date, end=end_date, api_key=ALPHAVANTAGE_API_KEY)
tesla_df.insert(5, 'stock', 'TSLA')

ibm_df = pdr.data.DataReader('IBM', 'av-daily', start=start_date, end=end_date, api_key=ALPHAVANTAGE_API_KEY)
ibm_df.insert(5, 'stock', 'IBM')

microsoft_df = pdr.data.DataReader('MSFT', 'av-daily', start=start_date, end=end_date, api_key=ALPHAVANTAGE_API_KEY)
microsoft_df.insert(5, 'stock', 'MSFT')
```

### Step 7. Combine the four separate dataFrames into one combined dataFrame df that holds the information for all four stocks

```python
combined_df = pd.concat([apple_df, tesla_df, ibm_df, microsoft_df])
combined_df
```

### Step 8. Shift the stock column into the index (making it a multi-level index consisting of the ticker symbol and the date).

```python
combined_df.index.name = 'date'
combined_df.reset_index(inplace=True)
combined_df.set_index(['stock', 'date'], inplace=True)
combined_df
```

### Step 7. Create a dataFrame called vol, with the volume values.

```python
vol = pd.DataFrame(combined_df['volume'])
vol
```

### Step 8. Aggregate the data of volume to weekly.


Hint: Be careful to not sum data from the same week of 2015 and other years.

```python
vol1 = vol.reset_index()
vol1.date = pd.to_datetime(vol1.date)
vol2 = vol1.set_index(['date'])
vol3 = vol2.groupby('stock').resample('W').sum()
vol3
```

### Step 9. Find all the volume traded in the year of 2015

```python
vol_total = vol2.resample('A').sum()
vol2015 = vol_total.iloc[0]
vol2015
```
